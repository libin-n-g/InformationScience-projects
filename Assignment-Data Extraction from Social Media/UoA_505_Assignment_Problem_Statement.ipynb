Getting video data using API for "avatar movie" search string. 

```python
from googleapiclient.discovery import build
with build('youtube', 'v3', developerKey=API_KEY) as youtube_service:
    request = youtube_service.search().list(
        part="snippet",
        q="avatar movie",
        regionCode="US",
        order="relevance",
        maxResults=50,
        type="video",
    )
    response = request.execute()
```

```python
def get_video_data(id_keys, part="snippet", **kwargs):
    with build('youtube', 'v3', developerKey=API_KEY) as youtube_service:
        request = youtube_service.videos().list(
            part=part,
            id=",".join(id_keys),
            **kwargs
        )
        response_for_video = request.execute()
    return response_for_video
```

```python
first_item_key = response["items"][0]["id"]["videoId"]
video_data = get_video_data([first_item_key])
```

```python

 # ID, Snippet, Channel ID, Video Description, Channel Title, Video Title
def get_data_from_video_snippet(snippet):
    data = {}
    data["Channel ID"] = snippet["channelId"].strip()
    data["Video Description"] = snippet["description"].strip()
    data["Channel Title"] = snippet["channelTitle"].strip()
    data["Video Title"] = snippet["title"].strip()
    return data
full_data = []
for item in response["items"]:
    data = {}
    data["ID"] = item["id"]["videoId"].strip()
    data["Snippet"] = item["snippet"]
    data.update(get_data_from_video_snippet(item["snippet"]))
    full_data.append(data)
avatar_movie_df = pd.DataFrame(full_data)
avatar_movie_df
```


> 2.b  Provide the following statistics for query string “avatar movie” of top 50 videos sorted by relevance in the US region ( 7 marks )

> Output expected: video ID, title, no of views, no of likes,no of comments exported to CSV file






Reference link: https://developers.google.com/youtube/v3/docs/videos/list

```python
video_response = get_video_data(avatar_movie_df["ID"], part="snippet,statistics")
def get_data_from_video_statictics(statistics):
    return {
        "number_of_views": int(statistics["viewCount"]),
        "number_of_likes": int(statistics["likeCount"]),
        "number_of_favourates": int(statistics["favoriteCount"]),
        "number_of_comments": int(statistics["commentCount"]),
    }
video_with_statistics_output = []
for item in video_response["items"]:
    data = {}
    data["ID"] = str(item["id"])
    data_statistics = get_data_from_video_statictics(item["statistics"])
    data_snippet = get_data_from_video_snippet(item["snippet"])
    data["Title"] = str(data_snippet["Video Title"])
    data["Number of Views"] = data_statistics["number_of_views"]
    data["Number of Likes"] = data_statistics["number_of_likes"]
    data["Number of Comments"] = data_statistics["number_of_comments"]
    video_with_statistics_output.append(data)
video_with_statistics_output_pd = pd.DataFrame(video_with_statistics_output)
video_with_statistics_output_pd.to_csv("avatar_movie_statistics.csv", index=False)
video_with_statistics_output_pd
```
